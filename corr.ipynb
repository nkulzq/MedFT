```python
import json
import matplotlib.pyplot as plt
```

```python
log = []
with open("./output/tta_decoder_official/log.txt", 'r') as f:
    for line in f:
        line = line.strip()
        log_entry = json.loads(line)
        log.append(log_entry)
```

```python
train_loss = [float(item['train_loss']) for item in log][:50]
epochs = range(1, len(train_loss) + 1)
plt.plot(epochs, train_loss, marker='o')
plt.xlabel('epoch')
plt.ylabel('loss')
plt.show()
```

```python
train_loss = [1 - float(item['train_loss']) for item in log][50:]
results = json.load(open("tta_decoder_scores.json", 'r'))
scores = [item['Bleu'][0] for item in results][50:]
plt.scatter(x=train_loss, y=scores, s=15)
plt.xlabel('train silhoette')
plt.ylabel('test BLEU0')
```

```python
from scipy import stats
stats.spearmanr(train_loss, scores)
```
